Zestaw danych zawiera dane skupiające się na danych osób cierpiących na cukrzycę oraz inne zależości jak wiek, ciąże, poziom glukozy, insuliny, czy też BMI osoby. 

```python
#Import niezbędnych bibliotek
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from random import randint
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold

df = pd.read_csv('diabetes2.csv') #Plik .csv z danymi

zakres = np.ptp(df['Age']) #Różnica min/max wartości kolumny
average = np.mean(df['Age']) #Średnia
median = np.median(df['Age']) #Mediana
wariancja = np.var(df['Age']) #Wariancja

print("zakres = " + str(zakres))
print("średnia = " + str(average))
print("mediana = " + str(median))
print("wariancja = " + str(wariancja))
```

```python
#Zilustrowanie zależności ciąż a wieku osoby, biorąc pod uwagę cukrzycę (żółty)

plt.scatter(df['Pregnancies'], df['Age'], c=df['Outcome'])
plt.xlabel('Pregnancies')
plt.ylabel('Age')
```

```python
#Zilustrowanie zależności poziomu glukozy a insuliny, biorąc pod uwagę cukrzycę (żółty)

plt.scatter(df['Glucose'], df['Insulin'], c=df['Outcome'])
plt.xlabel('Glucose')
plt.ylabel('Insulin')

X = df[['Glucose','Insulin']].values
y = df['Outcome'].values

model = LogisticRegression()
model.fit(X, y)
print(model.coef_, model.intercept_) #Ilustruje optymalny model liniowy
```

```python
#Zilustrowanie zależności ciśnienia krwi a BMI osoby, biorąc pod uwagę cukrzycę (żółty)

plt.scatter(df['BloodPressure'], df['BMI'], c=df['Outcome'])
plt.xlabel('BloodPressure')
plt.ylabel('BMI')
```

```python
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'DiabetesPedigreeFunction', 'Age']].values
y = df['Outcome'].values

model = LogisticRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("accuracy:", accuracy_score(y, y_pred)) #Dokładność
print("precision:", precision_score(y, y_pred)) #Precyzja
print("recall:", recall_score(y, y_pred)) #Pełność
print("f1 score:", f1_score(y, y_pred)) #Wynik F1
print(confusion_matrix(y, y_pred)) #Macierz pomyłek

print("Prawidzie negatywna (TN): 442")
print("Fałszywie pozytywna (FP): 58")
print("Fałszywie negatywna (FN): 113")
print("Prawdziwie pozytwyna (TP): 155")
```

```python
#funkcja, która pozwala nam przeprowadzić K-krotny sprawdzian testowy.
#Funkcja pozwalająca na K-krotny sprawdzian testowy.
#Argumenty to: 
# x - kolumny na podstawie, określenie celu
# y - nasz cel
# n_kf - krotność sprawdzianu

def score_model(X, y, n_kf):
    scores = []
    kf = KFold(n_splits=n_kf, shuffle=True)
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model = LogisticRegression()
        model.fit(X_train, y_train)
        scores.append(model.score(X_test, y_test))
    print(scores)
    print(np.mean(scores))

score_model(X, y, 3)
```

```python
#Sprawdzenie funkcji dla innych danych wejściowych

X_1 = df[['Pregnancies', 'BMI', 'Age']].values

score_model(X_1, y, 3)
```
